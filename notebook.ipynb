# Analysis

## The purpose of this analysis is to understand which method is most efficient in promoting a new product.

### For this test, 3 groups were made:

-  The ‘**Email**’ group, where the customers received and email during product launch and a follow up email three weeks later,

- The ‘**Call**’ group, where the customers received a call from the sales team that lasted about 30’ per customer on average,

- And lastly the ‘**Email + Call**’ group, where the customers first received an email then were called a week later by the sales team to discuss possible needs. The call in this scenario lasted about 10’ per customer on average.

### To understand which method is the most profitable, we must first find:

- What is the number of customers in each approach?

- What does the revenue look like overall and by each method?

- And if there was any difference over time for each method.

It would seem that the ‘Email’ group was the most efficient method since it brought in 50% of the total revenue, however after taking a closer look we can see that there is another group that seems a bit promising, so before we make our final conclusion, let's see how each group performed over time.

If we look at the revenue accross our testing period we see that it fluctuates.

![Revenue Across Week After Product Launch](Revenue%20Across%20Week%20After%20Product%20Launch.png)

To better understand this behaviour we must dig a little bit deeper and see how the revenue looks like by each method.

![Revenue by Method Across Week After Product Launch](Revenue%20by%20Method%20Across%20Week%20After%20Product%20Launch.png)

We can see that the revenue of the ‘**Email**’ group had a very strong first week with a revenue of about 84% of the week's total revenue, which as a result created a strong first week in total, however it doesn’t maintain its profit numbers which drop by every week.

For the ‘**Call**’ group the revenue seems to not take off, keeping a relatively consistent revenue with a slight increase as the weeks go by.

Lastly, the revenue of the ‘**Email + Call**’ group however seems to have the opposite effect compared to the ‘**Email**’ group, by having its first week being the weakest and increasing over time being the most profitable method out of the three after the fourth week.

So far we have seen how the total revenue looks like by each method used, but what would we get if we were interested in the  purchase made by each customer. 

![Number of Sales by Method](Number%20of%20Sales%20by%20Method.png)

We can see that the ‘**Email + Call**’ group has a higher median sales count and a larger volume compared to the other two groups. Also seeing a narrower variability (IQR) indicates a more consistent performance.

![Sales Revenue by Method](Sales%20Revenue%20by%20Method.png)

We can also see that in terms of revenue the customers that received the ‘**Email + Call**’ treatment dominated with a higher median revenue and higher revenue outliers making it the most profitable strategy per customer. The ‘**Email**’ and ‘**Call**’ group on the other hand are showing a lower and more variable revenue distribution.

![Engagement by Method](Engagement%20by%20Method.png)

Lastly, if we take a look at the number of engagement, we see that the ‘**Email + Call**’ group has a slightly higher median compared to the other two methods. 

Now, what information do we get from this graph ? 

Even though the ‘**Email**’ group has a higher customer count, the median number of sales and median sales per customer are not the highest of the three groups. 

In the ‘**Call**’ group the average sales are the lowest among the three groups with the exception of a few outliers.
